```python
def maxentry(triplelist):
    maxlist = []
    for entry in triplelist:
        maxvalue = max(entry)
        maxlist.append(maxvalue)
    
    return maxlist


```

## Problem 2

ðŸŽ¯Write a function that takes no arguments, gets the current time, and returns to the caller the integer number of minutes left in the current day (til midnight ðŸŒƒ)

```python
import datetime

def mintilmidnight():
    current_time = datetime.datetime.now()
    current_time_minutes = (current_time.hour*60)+(current_time.minute)+(int(round(current_time.second/60)))
    minutes = 1440 - current_time_minutes
    return minutes


print(mintilmidnight())
```

```python
def mortgage_payments(principal, intrate, monthpay):
    month = []
    payment = []
    interest = []
    balance = []
    
    #define 1st month
    month.append(1)
    payment.append(monthpay)
    interest.append(principal*intrate/1200)
    balance.append(principal-monthpay+interest[0])
    balancecurrent = balance[0]
    
    #continue until balance is 0
    i = 1
    while balancecurrent > 0:
        month.append(i+1)
        interest.append(balance[i-1]*intrate/1200)
        if monthpay > balance[i-1]:
            payment.append(balance[i-1]+interest[i]) 
        else:
            payment.append(monthpay)
        if monthpay < balance[i-1]:
            balance.append(balance[i-1]-monthpay+interest[i])
        else:
            balance.append(0)
        if balance[i] > balance[i-1]:
            print('Error: Monthly payment too low.')
            break
        balancecurrent = balance[i]
        i += 1
    
    return month, payment, interest, balance;

def totalpayment(month, payment, balance):
    #print out answers
    years = month[-1]//12
    monthsleft = month[-1]%12
    totalpayment = sum(payment)
    if balance[1] < balance[0]:
        return years, monthsleft, totalpayment;
    else:
        print('Error: Monthly payment too low.')
        return;


def amortizationtable (month, payment, interest, balance):
#make amoritzation table
    if balance[1] < balance[0]:
        print("Month\t\tPayment\t\tInterest\tBalance")
        i = 0
        for i in range(month[-1]):
            print("{:.0f}\t\t{:.2f}\t\t{:.2f}\t\t{:.2f}".format(month[i],payment[i],interest[i], balance[i]))
    return;

month, payment, interest, balance = mortgage_payments(100000,5,500)
years, monthsleft, totalpayment = totalpayment(month, payment, balance)
print("It took {} years and {} months to pay a total of ${:.2f} in {} payments.".format(years, monthsleft,totalpayment, month[-1]))
amortizationtable(month,payment,interest,balance)
```
